## Location ID

## change the location ID to dummy variable specify uptown/downtown/midtown/other

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
import matplotlib.pyplot as plt

# Load your cleaned sample
df = pd.read_csv('sample_J.csv')

downtown = [
    261, 88, 87, 209, 231, 144, 125, 114, 79, 12, 113, 74, 24,
    45, 43, 4, 42, 41, 50, 140, 142, 143, 239
]
midtown = [
    230, 90, 100, 161, 163, 164, 186, 186, 161, 170, 162, 229,
    233, 137, 107, 232, 48, 68, 113, 114, 246, 166, 144, 74
]
uptown = [
    116, 224, 243, 120, 127, 128, 153, 151, 238, 41, 152, 74, 75,
    194, 202, 262, 236, 233
]
df['PU_is_midtown'] = df['PULocationID'].isin(midtown).astype(int)
df['PU_is_uptown'] = df['PULocationID'].isin(uptown).astype(int)
df['PU_is_downtown'] = df['PULocationID'].isin(downtown).astype(int)
df['DO_is_midtown'] = df['DOLocationID'].isin(midtown).astype(int)
df['DO_is_uptown'] = df['DOLocationID'].isin(uptown).astype(int)
df['DO_is_downtown'] = df['DOLocationID'].isin(downtown).astype(int)
df = df.drop(columns=["PULocationID", 'DOLocationID'])
```

```python
df.head()
```

## Convert vendor ID to dummy variables

```python
df = pd.get_dummies(df, columns=["VendorID"], prefix="vendor", drop_first=True)

```

```python
df['vendor_2'] = df['vendor_2'].astype(int)
```

```python
df.head()
```

## Convert Weekend/Weekday to dummies

```python
weekend = [5,6]
df['is_weekend'] = df['pickup_weekday'].isin(weekend).astype(int)
```

```python
df.head()
```

# Numerical here

```python
# Define numeric predictors and target
num_feats = [
    'passenger_count', 'trip_distance',
    'fare_amount', 'extra',
    'tolls_amount', 'improvement_surcharge',
    'congestion_surcharge', 'airport_fee'
]
cat = [
    'is_peak_hour', 'is_night_hour',
    'PU_is_midtown', 'PU_is_uptown', 'PU_is_downtown',
    'DO_is_midtown', 'DO_is_uptown', 'DO_is_downtown',
    'vendor_2', 'is_weekend','tip_amount'
]
X = df[num_feats]
y = df['tip_amount']

# Drop missing values
X = X.dropna()
y = y.loc[X.index]

# Standardize numeric columns
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit Lasso with 5â€‘fold CV
lasso = LassoCV(cv=5, max_iter=5000, random_state=42)
lasso.fit(X_scaled, y)

# Extract and sort coefficients
coef = pd.Series(lasso.coef_, index=num_feats).sort_values()
```

```python
# Convert standardized features back to DataFrame with column names
scaled_df = pd.DataFrame(X_scaled, columns=num_feats, index=X.index)
```

```python
X_cat = df.loc[df.index, cat]  # make sure the index matches
X_final = pd.concat([scaled_df, X_cat], axis=1)
```
